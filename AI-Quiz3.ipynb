```python
import networkx as nx
```

```python
nodes = [i for i in range(1,8)]
edges = [
    (1,2),(1,3),(1,4),(1,5),
    (2,3),(2,5),
    (3,4),(3,6),
    (4,5),(4,6),(4,7),
    (6,7)
]

G = nx.Graph()
G.add_nodes_from(nodes)
G.add_edges_from(edges)
```

```python
nx.draw(G, with_labels=True)
```

```python
coloring = nx.algorithms.coloring.greedy_color(G)

rev = {}
for key, val in coloring.items():
    if val in rev:
        rev[val].append(key)
    else:
        rev[val] = [key]
        
print(rev)
```
