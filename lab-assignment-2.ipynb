```python
# Artificial training data
import numpy as np

np.random.seed(42)
m = 200
X = np.random.rand(m, 1)
y = 4 * (X - 0.5) ** 2
Y = y + np.random.randn(m, 1) / 10
```

**Q1. Plot the training data**

```python
# Plot the data
import matplotlib.pyplot as plt

plt.scatter(X, Y)
```

**Q2. Construct the decision tree**

```python
# Construct a Decision Tree
from sklearn.tree import DecisionTreeRegressor

clf = DecisionTreeRegressor(max_depth=2, random_state=42)
clf.fit(X, Y)
```

**Q3. Visualise the Decision Tree**

```python
# Plot the regression tree
from sklearn.tree import plot_tree

tree_plot = plot_tree(clf)
```

**Q4. Plot the Decision Tree**

```python
# Make some predictions

X_test = np.reshape(np.linspace(0, 1, 50), (-1,1))
Y_test = clf.predict(X_test)

# Plot the prediction
plt.figure()
plt.scatter(X, Y, s=20, edgecolor="black", c="darkorange", label="data")
plt.plot(X_test, Y_test, color="cornflowerblue", label="prediction", linewidth=2)
plt.xlabel("data")
plt.ylabel("target")
plt.title("Decision Tree Regression")
plt.legend()
plt.show()
```

**Q5. Plot the decision boundries for the three tree models**.

The `Y` value is scaled up 1000x.

```python
# Plot decision tree region
from mlxtend.plotting import plot_decision_regions


def plot_surface(clf):
    fig, ax = plt.subplots(figsize=(15,3))
    Y_predict = clf.predict(X)
    Y_predict = (Y_predict * 1000).astype(int) # plot_decision_tree() only accepts catagorical outputs. Scale up the value
    plot_decision_regions(X, Y_predict, clf=clf)

    ax = plt.gca()
    ax.set_xlim([-0.1, 1.1])

    plt.xlabel('x')
    plt.ylabel('y')
    plt.legend(loc='upper left')
    plt.tight_layout()

    plt.show()
```

```python
# max_depth = 3
clf_3 = DecisionTreeRegressor(max_depth=3, random_state=42)
clf_3.fit(X, Y)

# max_depth = 4
clf_4 = DecisionTreeRegressor(max_depth=4, random_state=42)
clf_4.fit(X, Y)

# min_samples_leaf = 10
clf_10 = DecisionTreeRegressor(min_samples_leaf=10, random_state=42)
clf_10.fit(X, Y)
```

```python
plot_surface(clf_3)
```

```python
plot_surface(clf_4)
```

```python
plot_surface(clf_10)
```
